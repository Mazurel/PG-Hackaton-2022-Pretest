```python
import numpy as np
import pandas as pd
import io
```

```python
f = io.open("data", mode="r", encoding="utf-8")
data = f.read()
data = data.split('\n')
```

```python
df = pd.DataFrame({'col':data})
df.apply(lambda x: pd.api.types.infer_dtype(x.values))
df
```

```python
def censor(sentence = ""):
    new_sentence = ""
    
    for word in sentence.split():
        if word in data:
            new_sentence += '*'*len(word) + ' '
        else:
            new_sentence += word + ' '
            
    return new_sentence
```

```python
censor("chuj cię strzeli")
```
